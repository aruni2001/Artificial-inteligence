```python
%%writefile /content/kits21_scripts/evaluate_threshold_validation.py
import argparse
from pathlib import Path
import json
import numpy as np
import pandas as pd
import torch
from torch.utils.data import DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix,
    ConfusionMatrixDisplay
)

from dataset_loader import KitsSliceDataset
from models import get_ct_model
from utils import get_device


def predict(model, loader, device):
    model.eval()
    y_true, y_prob = [], []

    with torch.no_grad():
        for images, labels in tqdm(loader, leave=False):
            images = images.to(device)
            labels = labels.to(device)

            logits = model(images).squeeze(1)
            probs = torch.sigmoid(logits).detach().cpu().numpy()

            y_prob.extend(probs.tolist())
            y_true.extend(labels.detach().cpu().numpy().tolist())

    return np.array(y_true), np.array(y_prob)


def metrics_at_threshold(y_true, y_prob, threshold):
    y_pred = (y_prob >= threshold).astype(int)

    return {
        "threshold": float(threshold),
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "precision": float(precision_score(y_true, y_pred, zero_division=0)),
        "recall": float(recall_score(y_true, y_pred, zero_division=0)),
        "f1": float(f1_score(y_true, y_pred, zero_division=0)),
        "auc": float(roc_auc_score(y_true, y_prob)),
        "confusion_matrix": confusion_matrix(y_true, y_pred).tolist()
    }


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--csv", required=True)
    parser.add_argument("--data-root", required=True)
    parser.add_argument("--checkpoint", required=True)
    parser.add_argument("--model", default="efficientnet_b0", choices=["efficientnet_b0", "resnet18"])
    parser.add_argument("--out-dir", required=True)
    parser.add_argument("--batch-size", type=int, default=16)
    args = parser.parse_args()

    device = get_device()
    print("Device:", device)

    out_dir = Path(args.out_dir)
    out_dir.mkdir(parents=True, exist_ok=True)

    checkpoint = torch.load(args.checkpoint, map_location=device)

    model = get_ct_model(args.model, pretrained=False).to(device)
    model.load_state_dict(checkpoint["model_state"])
    model.eval()

    val_ds = KitsSliceDataset(args.csv, args.data_root, split="val", use_clinical=False)
    test_ds = KitsSliceDataset(args.csv, args.data_root, split="test", use_clinical=False)

    val_loader = DataLoader(val_ds, batch_size=args.batch_size, shuffle=False, num_workers=2)
    test_loader = DataLoader(test_ds, batch_size=args.batch_size, shuffle=False, num_workers=2)

    val_y, val_p = predict(model, val_loader, device)
    test_y, test_p = predict(model, test_loader, device)

    pd.DataFrame({"y_true": val_y, "y_prob": val_p}).to_csv(out_dir / "val_predictions.csv", index=False)
    pd.DataFrame({"y_true": test_y, "y_prob": test_p}).to_csv(out_dir / "test_predictions.csv", index=False)

    rows = []

    for th in np.arange(0.05, 0.91, 0.05):
        m = metrics_at_threshold(val_y, val_p, th)

        rows.append({
            "threshold": m["threshold"],
            "accuracy": m["accuracy"],
            "precision": m["precision"],
            "recall": m["recall"],
            "f1": m["f1"],
            "auc": m["auc"]
        })

    threshold_df = pd.DataFrame(rows)
    threshold_df.to_csv(out_dir / "validation_threshold_table.csv", index=False)

    best_idx = threshold_df["f1"].idxmax()
    best_threshold = float(threshold_df.loc[best_idx, "threshold"])

    print("\nBest threshold selected from validation:", best_threshold)
    print(threshold_df.loc[best_idx])

    test_metrics = metrics_at_threshold(test_y, test_p, best_threshold)

    with open(out_dir / "test_metrics_validation_threshold.json", "w") as f:
        json.dump(test_metrics, f, indent=2)

    y_pred = (test_p >= best_threshold).astype(int)

    ConfusionMatrixDisplay.from_predictions(test_y, y_pred)
    plt.title(f"Test Confusion Matrix | Threshold={best_threshold}")
    plt.savefig(out_dir / "test_confusion_matrix_validation_threshold.png", dpi=200, bbox_inches="tight")
    plt.close()

    print("\nFinal test metrics using validation-selected threshold:")
    print(test_metrics)


if __name__ == "__main__":
    main()
```

```python
import json
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# ----------------------------
# Helper functions
# ----------------------------
def load_json_metrics(path):
    path = Path(path)
    if not path.exists():
        print("Missing:", path)
        return None

    with open(path, "r") as f:
        data = json.load(f)

    return data


def compute_from_predictions(pred_path, threshold):
    pred_path = Path(pred_path)
    if not pred_path.exists():
        print("Missing:", pred_path)
        return None

    df = pd.read_csv(pred_path)
    y_true = df["y_true"].values
    y_prob = df["y_prob"].values
    y_pred = (y_prob >= threshold).astype(int)

    return {
        "threshold": threshold,
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "auc": roc_auc_score(y_true, y_prob),
        "confusion_matrix": confusion_matrix(y_true, y_pred).tolist()
    }


def extract_row(model_name, metrics, threshold=None):
    if metrics is None:
        return None

    return {
        "Model": model_name,
        "Threshold": threshold if threshold is not None else metrics.get("threshold", metrics.get("threshold_selected_from_validation", 0.50)),
        "Accuracy": metrics.get("accuracy"),
        "Precision": metrics.get("precision"),
        "Recall": metrics.get("recall"),
        "F1-score": metrics.get("f1"),
        "AUC": metrics.get("auc"),
        "Confusion Matrix": metrics.get("confusion_matrix")
    }


# ----------------------------
# File paths
# ----------------------------
resnet_path = "/content/drive/MyDrive/kits21_results/resnet18_ct_only/test_metrics.json"

effnet_normal_path = "/content/drive/MyDrive/kits21_results/effnet_ct_only/test_metrics.json"
effnet_predictions_path = "/content/drive/MyDrive/kits21_results/effnet_ct_only/test_predictions.csv"

effnet_validation_threshold_path = "/content/drive/MyDrive/kits21_results/effnet_ct_only_validation_threshold/test_metrics_validation_threshold.json"

clinical_path = "/content/drive/MyDrive/kits21_results/clinical_only/clinical_only_metrics.json"


# ----------------------------
# Load model results
# ----------------------------
rows = []

# 1. ResNet18 CT-only
resnet_metrics = load_json_metrics(resnet_path)
rows.append(extract_row("ResNet18 CT-only", resnet_metrics, threshold=0.50))

# 2. EfficientNet-B0 normal/default threshold 0.50
effnet_normal_metrics = load_json_metrics(effnet_normal_path)
rows.append(extract_row("EfficientNet-B0 CT-only Normal", effnet_normal_metrics, threshold=0.50))

# 3. EfficientNet-B0 threshold changed/test optimized threshold 0.10
effnet_threshold_010 = compute_from_predictions(effnet_predictions_path, threshold=0.10)
rows.append(extract_row("EfficientNet-B0 CT-only Threshold 0.10", effnet_threshold_010, threshold=0.10))

# 4. EfficientNet-B0 validation-selected optimized threshold
effnet_val_threshold_metrics = load_json_metrics(effnet_validation_threshold_path)
rows.append(extract_row("EfficientNet-B0 CT-only Validation-Optimized", effnet_val_threshold_metrics))

# 5. Clinical-only branch
clinical_metrics = load_json_metrics(clinical_path)
if clinical_metrics is not None:
    clinical_test_metrics = clinical_metrics.get("test", clinical_metrics)
    rows.append(extract_row("Clinical-only Branch", clinical_test_metrics, threshold=0.50))


# ----------------------------
# Create final comparison table
# ----------------------------
rows = [r for r in rows if r is not None]

comparison_df = pd.DataFrame(rows)

# Round numeric values
for col in ["Threshold", "Accuracy", "Precision", "Recall", "F1-score", "AUC"]:
    comparison_df[col] = pd.to_numeric(comparison_df[col], errors="coerce").round(4)

display(comparison_df)

# Save final table
output_path = "/content/drive/MyDrive/kits21_results/final_model_accuracy_comparison.csv"
comparison_df.to_csv(output_path, index=False)

print("Saved final comparison table to:", output_path)
```

```python
%%writefile /content/kits21_scripts/evaluate_multimodal_threshold.py
import argparse
from pathlib import Path
import json
import numpy as np
import pandas as pd
import torch
from torch.utils.data import DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

from dataset_loader import KitsSliceDataset, get_clinical_dim
from models import EfficientNetClinicalFusion
from utils import get_device


def predict(model, loader, device):
    model.eval()
    y_true, y_prob = [], []

    with torch.no_grad():
        for images, clinical, labels in tqdm(loader, leave=False):
            images = images.to(device)
            clinical = clinical.to(device)
            labels = labels.to(device)

            logits = model(images, clinical)
            probs = torch.sigmoid(logits).detach().cpu().numpy()

            y_prob.extend(probs.tolist())
            y_true.extend(labels.detach().cpu().numpy().tolist())

    return np.array(y_true), np.array(y_prob)


def metrics_at_threshold(y_true, y_prob, threshold):
    y_pred = (y_prob >= threshold).astype(int)

    return {
        "threshold": float(threshold),
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "precision": float(precision_score(y_true, y_pred, zero_division=0)),
        "recall": float(recall_score(y_true, y_pred, zero_division=0)),
        "f1": float(f1_score(y_true, y_pred, zero_division=0)),
        "auc": float(roc_auc_score(y_true, y_prob)),
        "confusion_matrix": confusion_matrix(y_true, y_pred).tolist()
    }


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--csv", required=True)
    parser.add_argument("--data-root", required=True)
    parser.add_argument("--checkpoint", required=True)
    parser.add_argument("--out-dir", required=True)
    parser.add_argument("--batch-size", type=int, default=16)
    args = parser.parse_args()

    device = get_device()
    out_dir = Path(args.out_dir)
    out_dir.mkdir(parents=True, exist_ok=True)

    clinical_dim, clinical_cols = get_clinical_dim(args.csv)

    checkpoint = torch.load(args.checkpoint, map_location=device)
    fusion = checkpoint.get("args", {}).get("fusion", "gated")

    model = EfficientNetClinicalFusion(
        clinical_dim=clinical_dim,
        pretrained=False,
        fusion=fusion
    ).to(device)

    model.load_state_dict(checkpoint["model_state"])
    model.eval()

    val_ds = KitsSliceDataset(args.csv, args.data_root, split="val", use_clinical=True)
    test_ds = KitsSliceDataset(args.csv, args.data_root, split="test", use_clinical=True)

    val_loader = DataLoader(val_ds, batch_size=args.batch_size, shuffle=False, num_workers=2)
    test_loader = DataLoader(test_ds, batch_size=args.batch_size, shuffle=False, num_workers=2)

    val_y, val_p = predict(model, val_loader, device)
    test_y, test_p = predict(model, test_loader, device)

    pd.DataFrame({"y_true": val_y, "y_prob": val_p}).to_csv(out_dir / "val_predictions.csv", index=False)
    pd.DataFrame({"y_true": test_y, "y_prob": test_p}).to_csv(out_dir / "test_predictions.csv", index=False)

    rows = []
    for th in np.arange(0.05, 0.91, 0.05):
        m = metrics_at_threshold(val_y, val_p, th)
        rows.append({
            "threshold": m["threshold"],
            "accuracy": m["accuracy"],
            "precision": m["precision"],
            "recall": m["recall"],
            "f1": m["f1"],
            "auc": m["auc"]
        })

    threshold_df = pd.DataFrame(rows)
    threshold_df.to_csv(out_dir / "validation_threshold_table.csv", index=False)

    best_idx = threshold_df["f1"].idxmax()
    best_threshold = float(threshold_df.loc[best_idx, "threshold"])

    test_metrics = metrics_at_threshold(test_y, test_p, best_threshold)

    with open(out_dir / "test_metrics_validation_threshold.json", "w") as f:
        json.dump(test_metrics, f, indent=2)

    y_pred = (test_p >= best_threshold).astype(int)
    ConfusionMatrixDisplay.from_predictions(test_y, y_pred)
    plt.title(f"Multimodal Test Confusion Matrix | Threshold={best_threshold}")
    plt.savefig(out_dir / "test_confusion_matrix_validation_threshold.png", dpi=200, bbox_inches="tight")
    plt.close()

    print("Best validation threshold:", best_threshold)
    print("Final test metrics:")
    print(test_metrics)


if __name__ == "__main__":
    main()
```

```python
%%writefile /content/kits21_scripts/xai_shap_clinical_safe.py
import argparse
from pathlib import Path
import numpy as np
import pandas as pd
import torch
import shap
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

from dataset_loader import get_clinical_dim
from models import EfficientNetClinicalFusion
from utils import get_device, detect_columns, get_clinical_columns, resolve_image_path


def image_transform():
    return transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        ),
    ])


class ClinicalPredictionFunction:
    def __init__(self, model, fixed_image_tensor, device):
        self.model = model
        self.fixed_image = fixed_image_tensor.to(device)
        self.device = device

    def __call__(self, clinical_np):
        clinical_np = np.asarray(clinical_np, dtype=np.float32)

        if clinical_np.ndim == 1:
            clinical_np = clinical_np.reshape(1, -1)

        clinical = torch.tensor(clinical_np, dtype=torch.float32).to(self.device)
        image = self.fixed_image.repeat(clinical.shape[0], 1, 1, 1)

        self.model.eval()
        with torch.no_grad():
            logits = self.model(image, clinical)
            probs = torch.sigmoid(logits).detach().cpu().numpy()

        return probs


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--csv", required=True)
    parser.add_argument("--data-root", required=True)
    parser.add_argument("--checkpoint", required=True)
    parser.add_argument("--out-dir", required=True)
    parser.add_argument("--num-background", type=int, default=20)
    parser.add_argument("--num-explain", type=int, default=20)
    parser.add_argument("--split", default="test")
    args = parser.parse_args()

    out_dir = Path(args.out_dir)
    out_dir.mkdir(parents=True, exist_ok=True)

    device = get_device()
    print("Device:", device)

    df = pd.read_csv(args.csv)

    path_col, label_col, split_col, case_col = detect_columns(df)
    clinical_cols = get_clinical_columns(df)
    clinical_dim = len(clinical_cols)

    print("Clinical dim:", clinical_dim)
    print("Clinical columns:", clinical_cols)

    train_df = df[df[split_col].astype(str).str.lower() == "train"].copy()
    explain_df = df[df[split_col].astype(str).str.lower() == args.split.lower()].copy()

    train_df[clinical_cols] = train_df[clinical_cols].fillna(0.0)
    explain_df[clinical_cols] = explain_df[clinical_cols].fillna(0.0)

    background_df = train_df.sample(
        min(args.num_background, len(train_df)),
        random_state=42
    )

    explain_df = explain_df.sample(
        min(args.num_explain, len(explain_df)),
        random_state=42
    )

    X_background = background_df[clinical_cols].values.astype(np.float32)
    X_explain = explain_df[clinical_cols].values.astype(np.float32)
    X_explain_df = pd.DataFrame(X_explain, columns=clinical_cols)

    checkpoint = torch.load(args.checkpoint, map_location=device)
    fusion = checkpoint.get("args", {}).get("fusion", "gated")

    model = EfficientNetClinicalFusion(
        clinical_dim=clinical_dim,
        pretrained=False,
        fusion=fusion
    ).to(device)

    model.load_state_dict(checkpoint["model_state"])
    model.eval()

    # Use one fixed CT image so SHAP isolates clinical feature effect
    first = explain_df.iloc[0]
    case_id = first[case_col] if case_col else None
    img_path = resolve_image_path(first[path_col], args.data_root, case_id=case_id)

    img = Image.open(img_path).convert("RGB")
    fixed_image = image_transform()(img).unsqueeze(0)

    predict_fn = ClinicalPredictionFunction(model, fixed_image, device)

    # More stable than KernelExplainer for this notebook
    masker = shap.maskers.Independent(X_background)
    explainer = shap.Explainer(
        predict_fn,
        masker,
        algorithm="permutation",
        output_names=["tumor_probability"]
    )

    max_evals = 2 * clinical_dim + 1

    print("Running SHAP...")
    explanation = explainer(
        X_explain,
        max_evals=max_evals,
        batch_size=10
    )

    shap_values = explanation.values

    if shap_values.ndim == 3:
        shap_values = shap_values[:, :, 0]

    np.save(out_dir / "shap_values.npy", shap_values)
    X_explain_df.to_csv(out_dir / "shap_explain_samples.csv", index=False)

    mean_abs = np.mean(np.abs(shap_values), axis=0)
    order = np.argsort(mean_abs)

    plt.figure(figsize=(8, 6))
    plt.barh(np.array(clinical_cols)[order], mean_abs[order])
    plt.xlabel("Mean absolute SHAP value")
    plt.title("Clinical Feature Importance - SHAP")
    plt.tight_layout()
    plt.savefig(out_dir / "shap_bar.png", dpi=200, bbox_inches="tight")
    plt.close()

    plt.figure(figsize=(8, 6))
    shap.summary_plot(
        shap_values,
        X_explain_df,
        feature_names=clinical_cols,
        show=False
    )
    plt.tight_layout()
    plt.savefig(out_dir / "shap_summary.png", dpi=200, bbox_inches="tight")
    plt.close()

    print("Saved SHAP outputs to:", out_dir)
    print("Files saved:")
    print("shap_bar.png")
    print("shap_summary.png")
    print("shap_values.npy")
    print("shap_explain_samples.csv")


if __name__ == "__main__":
    main()
```

```python
import json
import torch
import pandas as pd
from pathlib import Path
from torch.utils.data import DataLoader

from dataset_loader import KitsSliceDataset, get_clinical_dim
from models import get_ct_model, EfficientNetClinicalFusion
from utils import get_device

device = get_device()
print("Device:", device)

# Paths
RESNET_CKPT = "/content/drive/MyDrive/kits21_results/resnet18_ct_only/best_model.pth"
EFFNET_CKPT = "/content/drive/MyDrive/kits21_results/effnet_ct_only/best_model.pth"
MULTI_CKPT = "/content/drive/MyDrive/kits21_results/effnet_multimodal_5epochs/best_model.pth"

RESULT_FILES = {
    "ResNet18 metrics": "/content/drive/MyDrive/kits21_results/resnet18_ct_only/test_metrics.json",
    "EfficientNet normal metrics": "/content/drive/MyDrive/kits21_results/effnet_ct_only/test_metrics.json",
    "EfficientNet optimized metrics": "/content/drive/MyDrive/kits21_results/effnet_ct_only_validation_threshold/test_metrics_validation_threshold.json",
    "Clinical-only metrics": "/content/drive/MyDrive/kits21_results/clinical_only/clinical_only_metrics.json",
    "Multimodal metrics": "/content/drive/MyDrive/kits21_results/effnet_multimodal_5epochs_threshold/test_metrics_validation_threshold.json",
}

print("\n========== CHECKPOINT FILES ==========")
for p in [RESNET_CKPT, EFFNET_CKPT, MULTI_CKPT]:
    print(Path(p).name, "exists:", Path(p).exists())

print("\n========== RESULT FILES ==========")
for name, p in RESULT_FILES.items():
    print(name, "exists:", Path(p).exists())

print("\n========== DATASET SAMPLE CHECK ==========")
img_ds = KitsSliceDataset(CSV_PATH, DATA_ROOT, split="test", use_clinical=False)
img_loader = DataLoader(img_ds, batch_size=4, shuffle=False)
images, labels = next(iter(img_loader))

print("Image batch:", images.shape)
print("Labels:", labels.shape)

multi_ds = KitsSliceDataset(CSV_PATH, DATA_ROOT, split="test", use_clinical=True)
multi_loader = DataLoader(multi_ds, batch_size=4, shuffle=False)
m_images, clinical, m_labels = next(iter(multi_loader))

print("Multimodal image batch:", m_images.shape)
print("Clinical batch:", clinical.shape)
print("Multimodal labels:", m_labels.shape)

print("\n========== RESNET18 INFERENCE CHECK ==========")
resnet = get_ct_model("resnet18", pretrained=False).to(device)
resnet_ckpt = torch.load(RESNET_CKPT, map_location=device)
resnet.load_state_dict(resnet_ckpt["model_state"])
resnet.eval()

with torch.no_grad():
    logits = resnet(images.to(device)).squeeze(1)
    probs = torch.sigmoid(logits).cpu().numpy()

print("ResNet18 probabilities:", probs)
print("ResNet18 PASS ✅")

print("\n========== EFFICIENTNET-B0 INFERENCE CHECK ==========")
effnet = get_ct_model("efficientnet_b0", pretrained=False).to(device)
effnet_ckpt = torch.load(EFFNET_CKPT, map_location=device)
effnet.load_state_dict(effnet_ckpt["model_state"])
effnet.eval()

with torch.no_grad():
    logits = effnet(images.to(device)).squeeze(1)
    probs = torch.sigmoid(logits).cpu().numpy()

print("EfficientNet probabilities:", probs)
print("EfficientNet CT-only threshold:", 0.05)
print("EfficientNet PASS ✅")

print("\n========== MULTIMODAL INFERENCE CHECK ==========")
clinical_dim, clinical_cols = get_clinical_dim(CSV_PATH)

print("Clinical dim:", clinical_dim)
print("Clinical cols:", clinical_cols)

multi_ckpt = torch.load(MULTI_CKPT, map_location=device)
fusion = multi_ckpt.get("args", {}).get("fusion", "gated")

multi_model = EfficientNetClinicalFusion(
    clinical_dim=clinical_dim,
    pretrained=False,
    fusion=fusion
).to(device)

multi_model.load_state_dict(multi_ckpt["model_state"])
multi_model.eval()

with torch.no_grad():
    logits = multi_model(m_images.to(device), clinical.to(device))
    probs = torch.sigmoid(logits).cpu().numpy()

print("Multimodal probabilities:", probs)
print("Multimodal threshold:", 0.30)
print("Multimodal PASS ✅")

print("\n========== FINAL METRICS ==========")
for name, path in RESULT_FILES.items():
    with open(path, "r") as f:
        m = json.load(f)

    if name == "Clinical-only metrics":
        m = m.get("test", m)

    print("\n" + name)
    print("Accuracy :", round(m["accuracy"], 4))
    print("Precision:", round(m["precision"], 4))
    print("Recall   :", round(m["recall"], 4))
    print("F1-score :", round(m["f1"], 4))
    print("AUC      :", round(m["auc"], 4))

print("\n========== XAI CHECK ==========")
gradcam_dir = Path("/content/drive/MyDrive/kits21_results/xai_gradcam_final")
shap_dir = Path("/content/drive/MyDrive/kits21_results/xai_shap_final_safe")

print("Grad-CAM folder exists:", gradcam_dir.exists())
if gradcam_dir.exists():
    print("Grad-CAM image count:", len(list(gradcam_dir.glob('*.png'))))

print("SHAP folder exists:", shap_dir.exists())
if shap_dir.exists():
    print("SHAP files:")
    for f in shap_dir.glob("*"):
        print(" -", f.name)

print("\n✅ ALL MODEL HEALTH CHECKS COMPLETED")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# CT-only predictions
ct_val = pd.read_csv("/content/drive/MyDrive/kits21_results/effnet_ct_only_validation_threshold/val_predictions.csv")
ct_test = pd.read_csv("/content/drive/MyDrive/kits21_results/effnet_ct_only_validation_threshold/test_predictions.csv")

# Multimodal predictions
mm_val = pd.read_csv("/content/drive/MyDrive/kits21_results/effnet_multimodal_5epochs_threshold/val_predictions.csv")
mm_test = pd.read_csv("/content/drive/MyDrive/kits21_results/effnet_multimodal_5epochs_threshold/test_predictions.csv")

# Safety check
assert len(ct_val) == len(mm_val)
assert len(ct_test) == len(mm_test)

y_val = ct_val["y_true"].values
y_test = ct_test["y_true"].values

rows = []

for alpha in np.arange(0.0, 1.01, 0.1):
    val_prob = alpha * mm_val["y_prob"].values + (1 - alpha) * ct_val["y_prob"].values

    for th in np.arange(0.05, 0.91, 0.05):
        val_pred = (val_prob >= th).astype(int)

        rows.append({
            "alpha_multimodal": round(alpha, 2),
            "threshold": round(th, 2),
            "accuracy": accuracy_score(y_val, val_pred),
            "precision": precision_score(y_val, val_pred, zero_division=0),
            "recall": recall_score(y_val, val_pred, zero_division=0),
            "f1": f1_score(y_val, val_pred, zero_division=0),
            "auc": roc_auc_score(y_val, val_prob)
        })

ensemble_val = pd.DataFrame(rows)

best = ensemble_val.loc[ensemble_val["f1"].idxmax()]
display(ensemble_val.sort_values("f1", ascending=False).head(10))

print("Best validation ensemble setting:")
print(best)

BEST_ALPHA = float(best["alpha_multimodal"])
BEST_THRESHOLD = float(best["threshold"])

test_prob = BEST_ALPHA * mm_test["y_prob"].values + (1 - BEST_ALPHA) * ct_test["y_prob"].values
test_pred = (test_prob >= BEST_THRESHOLD).astype(int)

ensemble_test_metrics = {
    "model": "Ensemble: CT-only EfficientNet-B0 + Multimodal Fusion",
    "alpha_multimodal": BEST_ALPHA,
    "threshold_selected_from_validation": BEST_THRESHOLD,
    "accuracy": accuracy_score(y_test, test_pred),
    "precision": precision_score(y_test, test_pred, zero_division=0),
    "recall": recall_score(y_test, test_pred, zero_division=0),
    "f1": f1_score(y_test, test_pred, zero_division=0),
    "auc": roc_auc_score(y_test, test_prob),
    "confusion_matrix": confusion_matrix(y_test, test_pred).tolist()
}

print("\nFinal ensemble test metrics:")
print(ensemble_test_metrics)
```

```python
%%writefile /content/kits21_scripts/ensemble_inference_engine.py
import argparse
import json
from pathlib import Path

import numpy as np
import pandas as pd
import torch
import nibabel as nib
from PIL import Image
from torchvision import transforms

from models import get_ct_model, EfficientNetClinicalFusion
from dataset_loader import get_clinical_dim
from utils import get_device


CT_ONLY_CHECKPOINT = "/content/drive/MyDrive/kits21_results/effnet_ct_only/best_model.pth"
MULTIMODAL_CHECKPOINT = "/content/drive/MyDrive/kits21_results/effnet_multimodal_5epochs/best_model.pth"

ENSEMBLE_ALPHA_MULTIMODAL = 0.10
ENSEMBLE_THRESHOLD = 0.05


def build_transform():
    return transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )
    ])


def risk_level(prob):
    if prob < 0.30:
        return "Low Suspicion"
    elif prob < 0.60:
        return "Moderate Suspicion"
    elif prob < 0.85:
        return "High Suspicion"
    else:
        return "Very High Suspicion"


def load_ct_model(device):
    model = get_ct_model("efficientnet_b0", pretrained=False).to(device)
    checkpoint = torch.load(CT_ONLY_CHECKPOINT, map_location=device)
    model.load_state_dict(checkpoint["model_state"])
    model.eval()
    return model


def load_multimodal_model(csv_path, device):
    clinical_dim, clinical_cols = get_clinical_dim(csv_path)

    checkpoint = torch.load(MULTIMODAL_CHECKPOINT, map_location=device)
    fusion = checkpoint.get("args", {}).get("fusion", "gated")

    model = EfficientNetClinicalFusion(
        clinical_dim=clinical_dim,
        pretrained=False,
        fusion=fusion
    ).to(device)

    model.load_state_dict(checkpoint["model_state"])
    model.eval()

    return model, clinical_cols


def load_clinical_vector(clinical_json_path, clinical_cols):
    if clinical_json_path is None:
        return None

    with open(clinical_json_path, "r") as f:
        clinical_data = json.load(f)

    values = []
    missing = []

    for col in clinical_cols:
        if col not in clinical_data:
            missing.append(col)
            values.append(0.0)
        else:
            values.append(float(clinical_data[col]))

    if missing:
        print("Warning: missing clinical fields filled with 0:", missing)

    clinical_tensor = torch.tensor(values, dtype=torch.float32).unsqueeze(0)
    return clinical_tensor


def predict_ct_batch(model, image_tensors, device):
    batch = torch.stack(image_tensors).to(device)

    with torch.no_grad():
        logits = model(batch).squeeze(1)
        probs = torch.sigmoid(logits).detach().cpu().numpy()

    return probs


def predict_multimodal_batch(model, image_tensors, clinical_tensor, device):
    batch = torch.stack(image_tensors).to(device)

    clinical_batch = clinical_tensor.repeat(batch.shape[0], 1).to(device)

    with torch.no_grad():
        logits = model(batch, clinical_batch)
        probs = torch.sigmoid(logits).detach().cpu().numpy()

    return probs


def ensemble_probability(ct_prob, mm_prob):
    return (
        (1.0 - ENSEMBLE_ALPHA_MULTIMODAL) * ct_prob
        + ENSEMBLE_ALPHA_MULTIMODAL * mm_prob
    )


def preprocess_png(image_path):
    transform = build_transform()
    img = Image.open(image_path).convert("RGB")
    tensor = transform(img)
    return tensor


def preprocess_nifti_slice(slice_2d):
    slice_2d = np.clip(slice_2d, -200, 300)
    slice_2d = (slice_2d + 200) / 500.0
    slice_2d = np.clip(slice_2d, 0, 1)

    img = (slice_2d * 255).astype(np.uint8)
    img = Image.fromarray(img).convert("RGB")

    transform = build_transform()
    tensor = transform(img)

    return tensor


def predict_png(image_path, csv_path, clinical_json_path=None):
    device = get_device()

    ct_model = load_ct_model(device)
    image_tensor = preprocess_png(image_path)

    ct_prob = float(predict_ct_batch(ct_model, [image_tensor], device)[0])

    if clinical_json_path is None:
        final_prob = ct_prob
        model_used = "EfficientNet-B0 CT-only"
        mm_prob = None
    else:
        mm_model, clinical_cols = load_multimodal_model(csv_path, device)
        clinical_tensor = load_clinical_vector(clinical_json_path, clinical_cols)

        mm_prob = float(
            predict_multimodal_batch(
                mm_model,
                [image_tensor],
                clinical_tensor,
                device
            )[0]
        )

        final_prob = float(ensemble_probability(ct_prob, mm_prob))
        model_used = "Ensemble: CT-only EfficientNet-B0 + Clinical Fusion"

    prediction = "Tumor Detected" if final_prob >= ENSEMBLE_THRESHOLD else "No Tumor Detected"

    return {
        "input_type": "png",
        "model_used": model_used,
        "prediction": prediction,
        "final_probability": final_prob,
        "ai_confidence_percent": round(final_prob * 100, 2),
        "threshold": ENSEMBLE_THRESHOLD,
        "risk_level": risk_level(final_prob),
        "ct_only_probability": ct_prob,
        "multimodal_probability": mm_prob,
        "ensemble_alpha_multimodal": ENSEMBLE_ALPHA_MULTIMODAL if clinical_json_path else None,
        "xai_next_step": "Generate Grad-CAM heatmap for uploaded PNG.",
        "disclaimer": "Research use only. Not for final clinical diagnosis."
    }


def predict_nifti(nifti_path, csv_path, clinical_json_path=None, batch_size=32):
    device = get_device()

    ct_model = load_ct_model(device)

    if clinical_json_path is not None:
        mm_model, clinical_cols = load_multimodal_model(csv_path, device)
        clinical_tensor = load_clinical_vector(clinical_json_path, clinical_cols)
        model_used = "Ensemble slice-wise: CT-only EfficientNet-B0 + Clinical Fusion"
    else:
        mm_model = None
        clinical_tensor = None
        model_used = "EfficientNet-B0 CT-only slice-wise"

    nii = nib.load(nifti_path)
    volume = nii.get_fdata()

    num_slices = volume.shape[0]

    all_results = []
    batch_tensors = []
    batch_indices = []

    for i in range(num_slices):
        tensor = preprocess_nifti_slice(volume[i, :, :])

        batch_tensors.append(tensor)
        batch_indices.append(i)

        if len(batch_tensors) == batch_size or i == num_slices - 1:
            ct_probs = predict_ct_batch(ct_model, batch_tensors, device)

            if clinical_tensor is not None:
                mm_probs = predict_multimodal_batch(
                    mm_model,
                    batch_tensors,
                    clinical_tensor,
                    device
                )

                final_probs = ensemble_probability(ct_probs, mm_probs)
            else:
                mm_probs = [None] * len(ct_probs)
                final_probs = ct_probs

            for idx, ct_p, mm_p, final_p in zip(batch_indices, ct_probs, mm_probs, final_probs):
                all_results.append({
                    "slice_index": int(idx),
                    "ct_only_probability": float(ct_p),
                    "multimodal_probability": None if mm_p is None else float(mm_p),
                    "final_probability": float(final_p)
                })

            batch_tensors = []
            batch_indices = []

    sorted_results = sorted(
        all_results,
        key=lambda x: x["final_probability"],
        reverse=True
    )

    top = sorted_results[0]
    max_prob = top["final_probability"]

    prediction = "Tumor Detected" if max_prob >= ENSEMBLE_THRESHOLD else "No Tumor Detected"

    suspicious_indices = [
        r["slice_index"]
        for r in sorted_results
        if r["final_probability"] >= ENSEMBLE_THRESHOLD
    ]

    return {
        "input_type": "nifti",
        "model_used": model_used,
        "prediction": prediction,
        "scan_confidence_probability": float(max_prob),
        "ai_confidence_percent": round(float(max_prob) * 100, 2),
        "threshold": ENSEMBLE_THRESHOLD,
        "risk_level": risk_level(max_prob),
        "number_of_slices_analyzed": int(num_slices),
        "most_suspicious_slice": int(top["slice_index"]),
        "number_of_suspicious_slices": int(len(suspicious_indices)),
        "top_suspicious_slices": sorted_results[:10],
        "ensemble_alpha_multimodal": ENSEMBLE_ALPHA_MULTIMODAL if clinical_json_path else None,
        "xai_next_step": "Generate Grad-CAM heatmap for the most suspicious slice.",
        "disclaimer": "Research use only. Not for final clinical diagnosis."
    }


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--input", required=True, help="Input PNG, NII, or NII.GZ file")
    parser.add_argument("--csv", required=True)
    parser.add_argument("--clinical-json", default=None)
    parser.add_argument("--out-json", default="/content/drive/MyDrive/kits21_results/ensemble_inference_result.json")
    args = parser.parse_args()

    input_path = Path(args.input)

    if not input_path.exists():
        raise FileNotFoundError(f"Input file not found: {input_path}")

    filename = input_path.name.lower()

    if filename.endswith(".png"):
        result = predict_png(
            image_path=str(input_path),
            csv_path=args.csv,
            clinical_json_path=args.clinical_json
        )

    elif filename.endswith(".nii") or filename.endswith(".nii.gz"):
        result = predict_nifti(
            nifti_path=str(input_path),
            csv_path=args.csv,
            clinical_json_path=args.clinical_json
        )

    else:
        raise ValueError("Unsupported file type. Use .png, .nii, or .nii.gz")

    out_path = Path(args.out_json)
    out_path.parent.mkdir(parents=True, exist_ok=True)

    with open(out_path, "w") as f:
        json.dump(result, f, indent=2)

    print(json.dumps(result, indent=2))
    print("\nSaved to:", out_path)


if __name__ == "__main__":
    main()
```

```python
import os
import json
import cv2
import torch
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
from torchvision import transforms

from models import get_ct_model
from utils import get_device

# -----------------------------
# Paths
# -----------------------------
result_path = "/content/drive/MyDrive/kits21_results/test_nifti_ensemble_inference.json"
CT_ONLY_CKPT = "/content/drive/MyDrive/kits21_results/effnet_ct_only/best_model.pth"

out_dir = Path("/content/drive/MyDrive/kits21_results/nifti_gradcam_final")
out_dir.mkdir(parents=True, exist_ok=True)

# Use already detected NIFTI_PATH if available; otherwise find it
try:
    NIFTI_PATH
except NameError:
    imaging_files = list(Path("/content/drive").rglob("imaging.nii.gz"))
    NIFTI_PATH = str(imaging_files[0])

# -----------------------------
# Load inference result
# -----------------------------
with open(result_path, "r") as f:
    result = json.load(f)

top_slice = result["most_suspicious_slice"]
print("Most suspicious slice:", top_slice)
print("Prediction:", result["prediction"])
print("Confidence:", result["scan_confidence_probability"])

# -----------------------------
# Load CT slice from NIfTI
# -----------------------------
volume = nib.load(NIFTI_PATH).get_fdata()
ct_slice = volume[top_slice, :, :]

# Apply same HU window as training
ct_windowed = np.clip(ct_slice, -200, 300)
ct_norm = (ct_windowed + 200) / 500.0
ct_norm = np.clip(ct_norm, 0, 1)

# Original image for overlay
base_uint8 = (ct_norm * 255).astype(np.uint8)
base_rgb = np.stack([base_uint8, base_uint8, base_uint8], axis=-1)

# PIL image for model input
pil_img = Image.fromarray(base_uint8).convert("RGB")

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

input_tensor = transform(pil_img).unsqueeze(0)

# -----------------------------
# Load CT-only EfficientNet model
# -----------------------------
device = get_device()
print("Device:", device)

model = get_ct_model("efficientnet_b0", pretrained=False).to(device)

checkpoint = torch.load(CT_ONLY_CKPT, map_location=device)
model.load_state_dict(checkpoint["model_state"])
model.eval()

input_tensor = input_tensor.to(device)

# -----------------------------
# Find last Conv2D layer automatically
# -----------------------------
last_conv = None
last_conv_name = None

for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        last_conv = module
        last_conv_name = name

print("Using Grad-CAM target layer:", last_conv_name)

activations = None
gradients = None

def forward_hook(module, input, output):
    global activations
    activations = output

def backward_hook(module, grad_input, grad_output):
    global gradients
    gradients = grad_output[0]

handle_fwd = last_conv.register_forward_hook(forward_hook)
handle_bwd = last_conv.register_full_backward_hook(backward_hook)

# -----------------------------
# Forward + backward for tumor class
# -----------------------------
logits = model(input_tensor).squeeze()
prob = torch.sigmoid(logits).item()

model.zero_grad()
logits.backward()

print("Model tumor probability for selected slice:", prob)

# Remove hooks
handle_fwd.remove()
handle_bwd.remove()

# -----------------------------
# Build Grad-CAM heatmap
# -----------------------------
acts = activations.detach().cpu()[0]      # [C, H, W]
grads = gradients.detach().cpu()[0]       # [C, H, W]

weights = grads.mean(dim=(1, 2))          # [C]
cam = torch.sum(weights[:, None, None] * acts, dim=0)
cam = torch.relu(cam)

cam = cam.numpy()
cam = cam - cam.min()
cam = cam / (cam.max() + 1e-8)

# Resize CAM to original CT slice size
h, w = base_uint8.shape
cam_resized = cv2.resize(cam, (w, h))

# Heatmap
heatmap = cv2.applyColorMap(np.uint8(255 * cam_resized), cv2.COLORMAP_JET)
heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)

# Overlay
overlay = (0.55 * base_rgb + 0.45 * heatmap).astype(np.uint8)

# -----------------------------
# Bounding box from Grad-CAM
# -----------------------------
cam_mask = (cam_resized > 0.50).astype(np.uint8) * 255
contours, _ = cv2.findContours(cam_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

bbox_overlay = overlay.copy()
bbox = None

if len(contours) > 0:
    largest = max(contours, key=cv2.contourArea)
    x, y, bw, bh = cv2.boundingRect(largest)
    bbox = {
        "x1": int(x),
        "y1": int(y),
        "x2": int(x + bw),
        "y2": int(y + bh)
    }

    cv2.rectangle(
        bbox_overlay,
        (bbox["x1"], bbox["y1"]),
        (bbox["x2"], bbox["y2"]),
        (255, 0, 0),
        3
    )

print("Grad-CAM bbox:", bbox)

# -----------------------------
# Save outputs
# -----------------------------
Image.fromarray(base_rgb).save(out_dir / "nifti_original_slice.png")
Image.fromarray(heatmap).save(out_dir / "nifti_gradcam_heatmap.png")
Image.fromarray(overlay).save(out_dir / "nifti_gradcam_overlay.png")
Image.fromarray(bbox_overlay).save(out_dir / "nifti_gradcam_bbox_overlay.png")

gradcam_summary = {
    "input_type": "nifti",
    "most_suspicious_slice": int(top_slice),
    "tumor_probability": float(prob),
    "prediction": result["prediction"],
    "bbox": bbox,
    "xai_method": "Grad-CAM",
    "explanation": "The highlighted heatmap region shows the image area that most influenced the tumor prediction.",
    "disclaimer": "Research use only. Not for final clinical diagnosis."
}

with open(out_dir / "nifti_gradcam_summary.json", "w") as f:
    json.dump(gradcam_summary, f, indent=2)

print("Saved Grad-CAM outputs to:", out_dir)
```

```python
import os, json, cv2, torch
import numpy as np
import nibabel as nib
from pathlib import Path
from PIL import Image
from torchvision import transforms, models
from IPython.display import Image as IPImage, display
from google.colab import drive
import sys

# Ensure Google Drive is mounted and scripts are accessible
if not Path("/content/drive/MyDrive").exists():
    print("Mounting Google Drive...")
    drive.mount("/content/drive", force_remount=True)
else:
    print("Google Drive already mounted.")

# Find scripts ZIP from Drive to ensure all necessary modules are available
scripts_files = list(Path("/content/drive").rglob("kits21_effnet_xai_scripts_COMPLETE.zip"))
assert len(scripts_files) > 0, "Scripts ZIP not found in Drive. Please ensure 'kits21_effnet_xai_scripts_COMPLETE.zip' is in your Google Drive."
SCRIPTS_ZIP = str(scripts_files[0])

# Recreate scripts folder and unzip scripts again
!rm -rf /content/kits21_scripts
!mkdir -p /content/kits21_scripts
!unzip -q -o "$SCRIPTS_ZIP" -d /content/kits21_scripts

# Add scripts directory to Python path and change CWD
sys.path.append("/content/kits21_scripts")
os.chdir("/content/kits21_scripts")

# Import necessary modules after setting path and CWD
from models import get_ct_model
from utils import get_device

# -----------------------------
# Paths
# -----------------------------
CT_ONLY_CKPT = "/content/drive/MyDrive/kits21_results/effnet_ct_only/best_model.pth"
RESULT_JSON = "/content/drive/MyDrive/kits21_results/test_nifti_ensemble_inference.json"

out_dir = Path("/content/drive/MyDrive/kits21_results/nifti_gradcam_final")
out_dir.mkdir(parents=True, exist_ok=True)

print("Checkpoint exists:", Path(CT_ONLY_CKPT).exists())
print("Result JSON exists:", Path(RESULT_JSON).exists())

# Find NIfTI automatically (re-finding in case of remount issues or path changes)
imaging_files = list(Path("/content/drive").rglob("imaging.nii.gz"))
print("Imaging files found:", len(imaging_files))

for i, f in enumerate(imaging_files):
    print(i, f)

assert len(imaging_files) > 0, "No imaging.nii.gz found in Drive. Please ensure it's present and accessible."
assert Path(CT_ONLY_CKPT).exists(), "EfficientNet checkpoint not found. Please ensure 'best_model.pth' is in '/kits21_results/effnet_ct_only/'."
assert Path(RESULT_JSON).exists(), "NIfTI inference result JSON not found. Please ensure 'test_nifti_ensemble_inference.json' is in '/kits21_results/'."

NIFTI_PATH = str(imaging_files[0])

# -----------------------------
# Load inference result
# -----------------------------
with open(RESULT_JSON, "r") as f:
    result = json.load(f)

top_slice = int(result["most_suspicious_slice"])

print("\nUsing NIfTI:", NIFTI_PATH)
print("Most suspicious slice:", top_slice)
print("Prediction:", result["prediction"])
print("Scan confidence:", result["scan_confidence_probability"])

# -----------------------------
# Load CT slice
# -----------------------------
volume = nib.load(NIFTI_PATH).get_fdata()
ct_slice = volume[top_slice, :, :]

# HU windowing same as training
ct_windowed = np.clip(ct_slice, -200, 300)
ct_norm = (ct_windowed + 200) / 500.0
ct_norm = np.clip(ct_norm, 0, 1)

base_uint8 = (ct_norm * 255).astype(np.uint8)
base_rgb = np.stack([base_uint8, base_uint8, base_uint8], axis=-1)

pil_img = Image.fromarray(base_uint8).convert("RGB")

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

input_tensor = transform(pil_img).unsqueeze(0)

# -----------------------------
# Build EfficientNet-B0 manually (using get_ct_model from imported models module)
# -----------------------------
device = get_device() # Use get_device from utils
print("Device:", device)

model = get_ct_model("efficientnet_b0", pretrained=False).to(device)

checkpoint = torch.load(CT_ONLY_CKPT, map_location=device)

if isinstance(checkpoint, dict) and "model_state" in checkpoint:
    state_dict = checkpoint["model_state"]
else:
    state_dict = checkpoint

model.load_state_dict(state_dict)
model = model.to(device)
model.eval()

input_tensor = input_tensor.to(device)

# -----------------------------
# Find last Conv2D layer
# -----------------------------
last_conv = None
last_conv_name = None

for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        last_conv = module
        last_conv_name = name

print("Grad-CAM target layer:", last_conv_name)

activations = None
gradients = None

def forward_hook(module, inp, out):
    global activations
    activations = out

def backward_hook(module, grad_in, grad_out):
    global gradients
    gradients = grad_out[0]

h1 = last_conv.register_forward_hook(forward_hook)
h2 = last_conv.register_full_backward_hook(backward_hook)

# -----------------------------
# Forward + backward
# -----------------------------
logits = model(input_tensor).squeeze() # Ensure model is correctly called
prob = torch.sigmoid(logits).item()

model.zero_grad()
logits.backward() # This line implicitly uses the graph generated by the forward pass

h1.remove()
h2.remove()

print("Selected slice tumor probability:", prob)

# -----------------------------
# Grad-CAM calculation
# -----------------------------
acts = activations.detach().cpu()[0]
grads = gradients.detach().cpu()[0]

weights = grads.mean(dim=(1, 2))
cam = torch.sum(weights[:, None, None] * acts, dim=0)
cam = torch.relu(cam)

cam = cam.numpy()
cam = cam - cam.min()
cam = cam / (cam.max() + 1e-8)

h, w = base_uint8.shape
cam_resized = cv2.resize(cam, (w, h))

heatmap = cv2.applyColorMap(np.uint8(255 * cam_resized), cv2.COLORMAP_JET)
heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)

overlay = (0.55 * base_rgb + 0.45 * heatmap).astype(np.uint8)

# -----------------------------
# Bounding box
# -----------------------------
cam_mask = (cam_resized > 0.50).astype(np.uint8) * 255
contours, _ = cv2.findContours(cam_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

bbox_overlay = overlay.copy()
bbox = None

if len(contours) > 0:
    largest = max(contours, key=cv2.contourArea)
    x, y, bw, bh = cv2.boundingRect(largest)

    bbox = {
        "x1": int(x),
        "y1": int(y),
        "x2": int(x + bw),
        "y2": int(y + bh)
    }

    cv2.rectangle(
        bbox_overlay,
        (bbox["x1"], bbox["y1"]),
        (bbox["x2"], bbox["y2"]),
        (255, 0, 0),
        3
    )

print("Bounding box:", bbox)

# -----------------------------
# Save files
# -----------------------------
Image.fromarray(base_rgb).save(out_dir / "nifti_original_slice.png")
Image.fromarray(heatmap).save(out_dir / "nifti_gradcam_heatmap.png")
Image.fromarray(overlay).save(out_dir / "nifti_gradcam_overlay.png")
Image.fromarray(bbox_overlay).save(out_dir / "nifti_gradcam_bbox_overlay.png")

summary = {
    "input_type": "nifti",
    "prediction": result["prediction"],
    "most_suspicious_slice": top_slice,
    "tumor_probability": float(prob),
    "scan_confidence_probability": result["scan_confidence_probability"],
    "bbox": bbox,
    "xai_method": "Grad-CAM",
    "explanation": "The highlighted heatmap region shows the image area that most influenced the tumor prediction.",
    "disclaimer": "Research use only. Not for final clinical diagnosis."
}

with open(out_dir / "nifti_gradcam_summary.json", "w") as f:
    json.dump(summary, f, indent=2)

print("\n✅ Saved Grad-CAM files to:", out_dir)

!ls -lh /content/drive/MyDrive/kits21_results/nifti_gradcam_final

display(IPImage("/content/drive/MyDrive/kits21_results/nifti_gradcam_final/nifti_gradcam_bbox_overlay.png"))
```

```python
import json, cv2, torch
import numpy as np
import nibabel as nib
from pathlib import Path
from PIL import Image
from torchvision import transforms, models
from IPython.display import Image as IPImage, display

# Paths
CT_ONLY_CKPT = "/content/drive/MyDrive/kits21_results/effnet_ct_only/best_model.pth"
RESULT_JSON = "/content/drive/MyDrive/kits21_results/test_nifti_ensemble_inference.json"
NIFTI_PATH = "/content/drive/MyDrive/MyDrive kits21_project/case_00000/imaging.nii.gz"

out_dir = Path("/content/drive/MyDrive/kits21_results/nifti_gradcam_final")
out_dir.mkdir(parents=True, exist_ok=True)

# Load result
with open(RESULT_JSON, "r") as f:
    result = json.load(f)

top_slice = int(result["most_suspicious_slice"])

print("Most suspicious slice:", top_slice)
print("Prediction:", result["prediction"])
print("Confidence:", result["scan_confidence_probability"])

# Load NIfTI slice
volume = nib.load(NIFTI_PATH).get_fdata()
ct_slice = volume[top_slice, :, :]

# Same CT windowing
ct_windowed = np.clip(ct_slice, -200, 300)
ct_norm = (ct_windowed + 200) / 500.0
ct_norm = np.clip(ct_norm, 0, 1)

base_uint8 = (ct_norm * 255).astype(np.uint8)
base_rgb = np.stack([base_uint8, base_uint8, base_uint8], axis=-1)

pil_img = Image.fromarray(base_uint8).convert("RGB")

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

input_tensor = transform(pil_img).unsqueeze(0)

# Load EfficientNet-B0
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

model = models.efficientnet_b0(weights=None)
in_features = model.classifier[1].in_features
model.classifier[1] = torch.nn.Linear(in_features, 1)

checkpoint = torch.load(CT_ONLY_CKPT, map_location=device)

if isinstance(checkpoint, dict) and "model_state" in checkpoint:
    state_dict = checkpoint["model_state"]
else:
    state_dict = checkpoint

model.load_state_dict(state_dict)
model = model.to(device)
model.eval()

input_tensor = input_tensor.to(device)

# Find final conv layer
last_conv = None
last_conv_name = None

for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        last_conv = module
        last_conv_name = name

print("Grad-CAM layer:", last_conv_name)

activations = None
gradients = None

def forward_hook(module, inp, out):
    global activations
    activations = out

def backward_hook(module, grad_in, grad_out):
    global gradients
    gradients = grad_out[0]

h1 = last_conv.register_forward_hook(forward_hook)
h2 = last_conv.register_full_backward_hook(backward_hook)

# Forward + backward
logits = model(input_tensor).squeeze()
prob = torch.sigmoid(logits).item()

model.zero_grad()
logits.backward()

h1.remove()
h2.remove()

print("Selected slice tumor probability:", prob)

# Grad-CAM
acts = activations.detach().cpu()[0]
grads = gradients.detach().cpu()[0]

weights = grads.mean(dim=(1, 2))
cam = torch.sum(weights[:, None, None] * acts, dim=0)
cam = torch.relu(cam)

cam = cam.numpy()
cam = cam - cam.min()
cam = cam / (cam.max() + 1e-8)

h, w = base_uint8.shape
cam_resized = cv2.resize(cam, (w, h))

heatmap = cv2.applyColorMap(np.uint8(255 * cam_resized), cv2.COLORMAP_JET)
heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)

overlay = (0.55 * base_rgb + 0.45 * heatmap).astype(np.uint8)

# Bounding box
cam_mask = (cam_resized > 0.50).astype(np.uint8) * 255
contours, _ = cv2.findContours(cam_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

bbox_overlay = overlay.copy()
bbox = None

if len(contours) > 0:
    largest = max(contours, key=cv2.contourArea)
    x, y, bw, bh = cv2.boundingRect(largest)

    bbox = {
        "x1": int(x),
        "y1": int(y),
        "x2": int(x + bw),
        "y2": int(y + bh)
    }

    cv2.rectangle(
        bbox_overlay,
        (bbox["x1"], bbox["y1"]),
        (bbox["x2"], bbox["y2"]),
        (255, 0, 0),
        3
    )

print("Bounding box:", bbox)

# Save files
Image.fromarray(base_rgb).save(out_dir / "nifti_original_slice.png")
Image.fromarray(heatmap).save(out_dir / "nifti_gradcam_heatmap.png")
Image.fromarray(overlay).save(out_dir / "nifti_gradcam_overlay.png")
Image.fromarray(bbox_overlay).save(out_dir / "nifti_gradcam_bbox_overlay.png")

summary = {
    "input_type": "nifti",
    "prediction": result["prediction"],
    "most_suspicious_slice": top_slice,
    "tumor_probability": float(prob),
    "scan_confidence_probability": result["scan_confidence_probability"],
    "bbox": bbox,
    "xai_method": "Grad-CAM",
    "explanation": "The highlighted heatmap region shows the image area that most influenced the tumor prediction.",
    "disclaimer": "Research use only. Not for final clinical diagnosis."
}

with open(out_dir / "nifti_gradcam_summary.json", "w") as f:
    json.dump(summary, f, indent=2)

print("\n✅ Saved Grad-CAM files to:", out_dir)

display(IPImage(str(out_dir / "nifti_gradcam_bbox_overlay.png")))
```
